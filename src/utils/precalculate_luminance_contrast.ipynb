```python
import os
import numpy as np    
import pandas as pd
import matplotlib.pyplot as plt             
import cv2
import sys
sys.path.append("..")
from main_classes.DataRetriever import load_npz
```

```python
def calculate_LC(img, kernel_size=40):
    LC_img = np.zeros(img.shape)
    mean_luminance = img.mean()
    for i in range(kernel_size//2, len(img)-kernel_size//2):
        for j in range(kernel_size//2, img.shape[-1]-kernel_size//2):
            mask = img[i-kernel_size//2:i+kernel_size//2,j-kernel_size//2:j+kernel_size//2]
            LC = mask.std()/mean_luminance
            LC_img[i, j] = LC
    return LC_img
```

```python
IMGS_FOLDER = '../../../data/images/'
```

```python
LC_IMGS_FOLDER = '../../../data/LC_images/'
```

```python
if not os.path.exists(LC_IMGS_FOLDER):
    os.makedirs(LC_IMGS_FOLDER)
for img_name in filter(lambda x: x.endswith(".png"), os.listdir(IMGS_FOLDER)):
    if img_name in os.listdir(LC_IMGS_FOLDER):
        continue
    print(img_name)
    img = cv2.imread(os.path.join(IMGS_FOLDER, img_name), 0)
    LC_img = calculate_LC(img, kernel_size=40)
    cv2.imwrite(os.path.join(LC_IMGS_FOLDER, img_name), LC_img*255)
```
