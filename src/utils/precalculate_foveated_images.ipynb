```python
if not os.path.exists(FOVEATED_IMGS_FOLDER):
    os.makedirs(FOVEATED_IMGS_FOLDER)
for img_name in filter(lambda x: x.endswith(".png"), os.listdir(IMGS_FOLDER)):
    print(f"Working on: {img_name}")
    save_filepath = os.path.join(FOVEATED_IMGS_FOLDER, img_name)
    if not os.path.exists(save_filepath):
        os.makedirs(save_filepath)
    img = cv2.imread(os.path.join(IMGS_FOLDER, img_name))
    if os.listdir(save_filepath):
        latest_x = max(list(map(lambda x: int(x.split("_")[0]), os.listdir(save_filepath))))
    else:
        latest_x = 0
    for x in range(latest_x, img.shape[1]):
        print(f"Working on x={x}")
        for y in range(img.shape[0]):
            save_img_name = f"{x}_{y}.png"
            if save_img_name in os.listdir(save_filepath):
                continue
            fov_img = foveat_img(img, fixs=[(x, y)])
            cv2.imwrite(os.path.join(save_filepath, save_img_name), fov_img*255)
```
